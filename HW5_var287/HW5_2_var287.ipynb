```python
SQL_SOURCE = 'https://fb55.carto.com:443/api/v2/sql?q='

import urllib2
import urllib
import StringIO
import ast
import pandas as pd

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()
```

```python
def get_data(test):
    try:
        return pd.read_csv(StringIO.StringIO(queryCartoDB(test)), sep=',')
    except ValueError as v:
        print (str(v))
```

```python
## Sort data by start_station_id and tripduration where tripduration is less than 3 hours
test  = '''
SELECT * 
FROM fb55.citibike 
WHERE tripduration < 10800
ORDER BY start_station_id, tripduration
LIMIT 10
;
'''

table = get_data(test)
table
```

```python
## Show last 10 of sorted data by start_station_id and tripduration where tripduration is less than 3 hours

test  = '''
SELECT * 
FROM fb55.citibike 
WHERE tripduration < 10800
ORDER BY start_station_id DESC, tripduration DESC
LIMIT 10
;
'''

table = get_data(test)
table
```

```python
## all unique start_station_id
test =  '''
SELECT DISTINCT start_station_id  as id
FROM fb55.citibike 
WHERE start_station_id IS NOT NULL;
'''

table = get_data(test)
table.id.unique()
```

```python
## count number of trips, avg, min, and max of trio duration
test = '''
SELECT MAX(tripduration) as Max, MIN(tripduration) as MIN, AVG(tripduration) as AVG, COUNT(tripduration) as COUNT
FROM fb55.citibike 
;
'''

table = get_data(test)
table
```

### Task 2

```python
## select trips started on feb 2, 2015 only
test = '''
SELECT *
FROM fb55.citibike
WHERE EXTRACT(MONTH FROM starttime::date) = 2 AND EXTRACT(DAY FROM starttime::date) = 1 AND EXTRACT(YEAR FROM starttime::date) = 2015

; 
'''

table = get_data(test)
table
```

```python
## average trip duration on weekend
test = '''
SELECT AVG(tripduration)
FROM fb55.citibike
WHERE EXTRACT(DAY FROM starttime::timestamp) = 7 OR EXTRACT(DAY FROM starttime::timestamp) = 1
; 
'''

table = get_data(test)
table
```

```python
## average trip duration on weekday
test = '''
SELECT AVG(tripduration)
FROM fb55.citibike
WHERE EXTRACT(DAY FROM starttime::timestamp) != 7 AND EXTRACT(DAY FROM starttime::timestamp) !=  1
; 
'''

table = get_data(test)
table
```
