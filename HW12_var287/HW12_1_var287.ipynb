Note on preprocessing: the preparation of these data should include the "standardization" of the time seris. Always when you apply a moachine learning model you should standardize the data or whiten it. These operations may be complicated preprocessing operation, they can for example invoke PCA, and they are aimed at removing covariance in the feature space.

The simplest version of whitening standardizes each feature" from every column (feature) you subtract the mean and you divide it by the standard deviation.

But the time series preprocessing is different if you want to use each time stamp as feature: because you want to maintain the coherence of the data, i.e. you **do not want** to treat each feature as an independent feature! the pattenrs in the time series are what you are clustering on (time series goes up and down, creating a pattern) and those patterns would be lost by whitening.

#### So when you cluster or otherwise use a time series in a model, using each timestamp as a feature, you want to standardize the observations, not the features.

From each row of your table (observation, in this case each zipcode) subtract the mean and you divide the row by the standard deviation.

## Data cleaning

```python
nycz_shape = gpd.GeoDataFrame.from_file(os.getenv("PUIDATA") + "/ZIP_CODE_040114.shp")

print("ZIPCODES:",
      len(np.unique(nycz_shape.ZIPCODE)), "POLYGONS:",
      len(nycz_shape.ZIPCODE),
      
      "the difference is due to island with same zip as parts of land,",
      "but disjoint shapes")

print("number of NYC zip codes:", len(nycz_shape.ZIPCODE.values.astype(int)))
```

```python
x1 = ['19' + str(i) for i in range(94,100)]
x2 = ['200' + str(i) for i in range(0,10)]
x3 = [ '20' + str(i) for i in range(10, 15)]

columns = x1 + x2 + x3
```

```python
zbpfiles = glob.glob('zbp*.zip')
```

```python
zipcode = [pd.read_csv(zipfile.ZipFile(fname).open(fname.replace('.zip', '.txt'))) for fname in zbpfiles]

for i in range(len(zbpfiles)):
    zipcode[i].columns = [x.lower() for x in zipcode[i].columns]
    
zipcodeS = [zipcode[['zip','est']] for zipcode in zipcode]

for i in range(len(zbpfiles)):
    zipcodeS[i] = zipcodeS[i].rename(columns={'est': columns[i]}) 

zipDF = reduce(lambda x, y: pd.merge(x, y, on = 'zip', how='inner'), zipcodeS)
```

```python
# get zip as index index

zipDF.index = zipDF['zip']
zipDF = zipDF.iloc[:,1:]
zipDF.head()
```

## Standardizing

```python
# standardzied using minmaxscaler learned from ADS
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
zipDF_scaled = pd.DataFrame(min_max_scaler.fit_transform(zipDF), columns=zipDF.columns, index=zipDF.index)
zipDF_scaled.head()
```

## Clustering 

```python
n_clusters = 4

km_Cluster = KMeans(n_clusters=n_clusters, random_state=444)

labels = km_Cluster.fit_predict(zipDF_scaled)
labels
```

```python
centroid =km_Cluster.cluster_centers_
centroid.shape
```

```python
zipDF_scaled['labels'] = labels
zipDF_scaled.head()
```

```python
for i in range(4):
    print('cluster', i , 'has ', len(zipDF_scaled[zipDF_scaled.labels == i]) , ' zipcodes')
```

```python
fig = plt.figure(figsize=(10,5))
fig.subplots_adjust(hspace=0.6, wspace=0.4)

ax = fig.add_subplot(2,2,1)
for i in range(len(zipDF[zipDF_scaled.labels == 0])):
    ax.plot(range(21), zipDF_scaled[zipDF_scaled.labels == 0].iloc[i,:-1].values, alpha=0.3)
    ax.plot(range(21), centroid[0], 'blue')
    
    ax.set_xlabel('Years')
    ax.set_ylabel('Establishments')
    ax.set_title('Cluster 1')
        
        
ax = fig.add_subplot(2,2,2)
for i in range(len(zipDF[zipDF_scaled.labels == 1])):
    ax.plot(range(21), zipDF_scaled[zipDF_scaled.labels == 1].iloc[i,:-1].values, alpha=0.3)
    ax.plot(range(21), centroid[1], 'blue')
    
    ax.set_xlabel('Years')
    ax.set_ylabel('Establishments')
    ax.set_title('Cluster 2')
    
ax = fig.add_subplot(2,2,3)
for i in range(len(zipDF[zipDF_scaled.labels == 2])):
    ax.plot(range(21), zipDF_scaled[zipDF_scaled.labels == 2].iloc[i,:-1].values, alpha=0.3)
    ax.plot(range(21), centroid[2], 'blue')
    
    ax.set_xlabel('Years')
    ax.set_ylabel('Establishments')
    ax.set_title('Cluster 3')
    
ax = fig.add_subplot(2,2,4)
for i in range(len(zipDF[zipDF_scaled.labels == 3])):
    ax.plot(range(21), zipDF_scaled[zipDF_scaled.labels == 3].iloc[i,:-1].values, alpha=0.3)
    ax.plot(range(21), centroid[3], 'blue')
    
    ax.set_xlabel('Years')
    ax.set_ylabel('Establishments')
    ax.set_title('Cluster 4')
```

Figure 1: Time Series graph of each clusters with respective centroid shown. It looks like the series in cluster 1, 3, and 4 are close to the centroid while cluster 2 could be clustered into more groups
